```python
import pandas as pd
import numpy as np
```

```python
orig_data = pd.DataFrame.from_csv('Clean_data/train2.csv', header=0)
orig = orig_data.ix[:, :'SYNERGY_SCORE']
drug_info = pd.DataFrame.from_csv('Drug_expanded/exp_imputed.csv', header=0)
```

```python
drug_info
```

```python
def generate_features(drug_info, orig):
    drug_props_A = drug_info.rename(columns=lambda x: 'A_'+x)
    drug_props_B = drug_info.rename(columns=lambda x: 'B_'+x)
    features = orig.join(drug_props_A, on='COMPOUND_A').join(drug_props_B, on='COMPOUND_B')
    return features
```

```python
df = generate_features(drug_info, orig)
df
```

```python
df.to_csv('Train_data/drug_exp2.csv')
```
