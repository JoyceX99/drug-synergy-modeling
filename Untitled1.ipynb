```python
import numpy as np
import pandas as pd
from sklearn import cross_validation, metrics
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.cluster import KMeans
from sklearn.preprocessing import Imputer
from scipy.stats import stats
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.DataFrame.from_csv('Train_data/with_interaction_biogrid.csv')
gex = pd.DataFrame.from_csv('Sanger_molecular_data/gex.csv', header=0).transpose()
gex.shape
```

```python
gex
```

```python
for col in gex.columns:    # for each gene:
    mean = np.mean(gex[col].values)
    std = np.std(gex[col].values)
    varies = ((gex[col] < mean-5*std) | (gex[col] > mean+5*std)).any() # any gene expression values 
                                                                                                 # more than 8 away from mean?
    if not varies:
        gex.drop(col, axis=1, inplace=True)
```

```python
gex.shape
```

```python
gex_costs = {}
for i in range(2, 15):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(gex.values)
    gex_costs[i] = (1/i) * kmeans.inertia_
gex_costs
```

```python
plt.plot([cluster for (cluster, cost) in gex_costs.items()],
         [cost for (cluster, cost) in gex_costs.items()], 'bo--')
plt.show()
```

```python
gex_kmeans = KMeans(n_clusters = 5, n_init=800)
gex_kmeans.fit(gex.values)
```

```python
gex_kmeans.labels_
```

```python
rows = df['CELL_LINE'].values
gex_to_cluster = gex.ix[rows]
gex_to_cluster.shape  
```

```python
imputer = Imputer(missing_values='NaN', strategy="mean", axis=0)
imputer.fit(gex_to_cluster)
gex_to_cluster = pd.DataFrame(imputer.transform(gex_to_cluster), columns = gex_to_cluster.columns, index= gex_to_cluster.index)
df['GEX_CLUSTER'] = gex_kmeans.predict(gex_to_cluster)
```

### Model

```python
df = df.iloc[np.random.permutation(len(df))]
y = df['SYNERGY_SCORE'].values
X = pd.get_dummies(df.drop('SYNERGY_SCORE', axis=1), columns=['CELL_LINE', 'COMPOUND_A', 'COMPOUND_B', 'Disease_area']).values
```

```python
def pearson(y, y_pred):
    return stats.pearsonr(y, y_pred)[0]
pearson_scorer = metrics.make_scorer(pearson)
```

```python
for i in range(3):
    gbr = GradientBoostingRegressor(n_estimators=1000, max_depth=5, loss='huber',
                                max_features='log2', learning_rate=0.03)
    pearsons = cross_validation.cross_val_score(gbr, X, y, scoring=pearson_scorer, cv=5)
    r2s = cross_validation.cross_val_score(gbr, X, y, cv=5)

    print(str(np.mean(r2s)) + ', ' + str(np.mean(pearsons)))
```
