```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
```

```python
orig_data = pd.DataFrame.from_csv('Clean_data/train.csv', header=0)
origX = orig_data.ix[:, :'Einf_B']
origY = orig_data.ix[:, 'SYNERGY_SCORE']
drug_info = pd.DataFrame.from_csv('Drug_synergy_data/Drug_info_release.csv', header=0, index_col=False)
```

```python
drug_info
```

```python
to_impute = drug_info.copy()
to_impute.drop(['ChallengeName', 'Target(Official Symbol)', 'SMILES or PubChem ID'], axis=1, inplace=True)
```

```python
# imputer = Imputer(missing_values='NaN', strategy="mean", axis=0)
# imputer.fit(to_impute)
# imputed = pd.DataFrame(imputer.transform(to_impute), columns = to_impute.columns)
imputed = pd.DataFrame.from_csv('Combined_data/drugInfo_imputed.csv')
```

```python
imputed
```

```python
full_imputed = pd.concat([drug_info[['ChallengeName', 'Target(Official Symbol)']], imputed], axis=1)
full_imputed.set_index('ChallengeName', inplace=True)
```

```python
full_imputed
```

```python
full_imputed_A = full_imputed.rename(columns=lambda x: 'A_'+x)
full_imputed_B = full_imputed.rename(columns=lambda x: 'B_'+x)
```

```python
mono_info = origX.join(full_imputed_A, on='COMPOUND_A').join(full_imputed_B, on='COMPOUND_B')
```

```python
mono_info
```

```python
mono_info.to_csv('Combined_data/drugInfo_combined.csv', index=False)
```

```python
# FOR NOW: drop target
mono_info.drop(['A_Target(Official Symbol)', 'B_Target(Official Symbol)'], axis=1, inplace=True)
```

```python
mono_info_int = pd.get_dummies(mono_info, columns=['CELL_LINE', 'COMPOUND_A', 'COMPOUND_B'])
```

# Algorithm Training

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation, preprocessing
```

```python
Xpd = mono_info_int
Xpd = Xpd.iloc[np.random.permutation(len(Xpd))]
Ypd = orig_data['SYNERGY_SCORE'].reindex(Xpd.index)
Xvals = Xpd.values
yvals = Ypd.values
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xvals, yvals)
```

## Random Forest

```python
rfr = RandomForestRegressor(max_features=0.25, max_leaf_nodes=None, 
                            min_samples_leaf = 1, max_depth = None, n_estimators = 50)
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

```python
n_features = Xvals.shape[1]
param_grid_rf = {
    'n_estimators': [30, 50, 80],
    'max_features': ['auto', 'sqrt', 0.25],
    'max_leaf_nodes': [None, n_features],
    'max_depth': [None, 10]
}
rfr = RandomForestRegressor()
```

```python
grid_search_rf = GridSearchCV(rfr, param_grid_rf, n_jobs=2)
for i in range(5):
    #X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xvals, yvals)
    grid_search_rf.fit(X_train, y_train)
    print(grid_search_rf.best_params_)
    print(grid_search_rf.best_score_)
```

## Adaboost

```python
param_grid_ab = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.5, 1, 2],
    'loss': ['linear', 'square', 'exponential']
}
abr = AdaBoostRegressor()
```

```python
grid_search_ab = GridSearchCV(abr, param_grid_ab, n_jobs=2)
for i in range(5):
    grid_search_ab.fit(X_train, y_train)
    print(grid_search_ab.best_params_)
    print(grid_search_ab.best_score_)
```

```python
abr = AdaBoostRegressor(loss='square', learning_rate=0.5)
abr.fit(X_train, y_train)
```

```python
abr.score(X_test, y_test)
```

## Gradient Boosting

```python
param_grid_gb = {
    'learning_rate': [0.03, 0.1],
    'n_estimators': [100, 150]
}
gbr = GradientBoostingRegressor()
```

```python
grid_search_gb = GridSearchCV(gbr, param_grid_gb, n_jobs=2)
for i in range(5):
    grid_search_gb.fit(X_train, y_train)
    print(grid_search_gb.best_params_)
    print(grid_search_gb.best_score_)
```

```python
gbr = GradientBoostingRegressor(loss='ls', n_estimators=100, learning_rate=0.2)
gbr.fit(X_train, y_train)
```

```python
gbr.score(X_test, y_test)
```

## SVR

```python
Xnorm = preprocessing.normalize(Xvals)
Ynorm = preprocessing.normalize(yvals).reshape(2145,)
X_train_norm, X_test_norm, y_train_norm, y_test_norm = cross_validation.train_test_split(Xnorm, Ynorm)
```

```python
param_grid_svr = {
    'C': [30, 90, 300],
    'epsilon': [0.3, 1, 3],
    'gamma': [0.0, 0.0015, 0.001]
}
svr = SVR()
```

```python
grid_search = GridSearchCV(svr, param_grid_svr, n_jobs=2)
for i in range(5):
    X_train_norm, X_test_norm, y_train_norm, y_test_norm = cross_validation.train_test_split(Xvals, yvals)
    grid_search.fit(X_train_norm, y_train_norm)
    print(grid_search.best_params_)
    print(grid_search.best_score_)
```
