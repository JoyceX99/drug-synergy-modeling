```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
```

```python
orig_data = pd.DataFrame.from_csv('Drug_synergy_data/ch1_train_combination_and_monoTherapy.csv', header=0, index_col=False)
origX = orig_data.ix[:, :'Einf_B']
origY = orig_data.ix[:, 'SYNERGY_SCORE']
drug_info = pd.DataFrame.from_csv('Drug_synergy_data/Drug_info_release.csv', header=0, index_col=False)
```

```python
drug_info
```

```python
to_impute = drug_info.copy()
to_impute.drop(['ChallengeName', 'Target(Official Symbol)', 'SMILES or PubChem ID'], axis=1, inplace=True)
```

```python
imputer = Imputer(missing_values='NaN', strategy="mean", axis=0)
imputer.fit(to_impute)
imputed = pd.DataFrame(imputer.transform(to_impute), columns = to_impute.columns)
```

```python
imputed
```

```python
full_imputed = pd.concat([drug_info[['ChallengeName', 'Target(Official Symbol)']], imputed], axis=1)
full_imputed.set_index('ChallengeName', inplace=True)
```

```python
full_imputed
```

```python
full_imputed_A = full_imputed.rename(columns=lambda x: 'A_'+x)
full_imputed_B = full_imputed.rename(columns=lambda x: 'B_'+x)
```

```python
mono_info = origX.join(full_imputed_A, on='COMPOUND_A').join(full_imputed_B, on='COMPOUND_B')
```

```python
mono_info
```

```python
# FOR NOW: drop target
mono_info.drop(['A_Target(Official Symbol)', 'B_Target(Official Symbol)'], axis=1, inplace=True)
```

```python
mono_info_int = pd.get_dummies(mono_info, columns=['CELL_LINE', 'COMPOUND_A', 'COMPOUND_B'])
```

# Algorithm Training

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation
```

```python
Xpd = mono_info_int
Ypd = orig_data['SYNERGY_SCORE']
Xvals = Xpd.values
yvals = Ypd.values
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xvals, yvals)
```

## Random Forest

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

```python
n_features = Xvals.shape[1]
param_grid_rf = {
    'n_estimators': [10, 20],
    'max_features': ['auto', 'sqrt', 0.25],
    'max_depth': [None, 5, 10],
    'min_samples_leaf': [1, 2],
    'max_leaf_nodes': [(int)(n_features/4), (int)(n_features/5)]
}
rfr = RandomForestRegressor()
```

```python
grid_search_rf = GridSearchCV(rfr, param_grid_rf, n_jobs=2)
for i in range(5):
    #X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xvals, yvals)
    grid_search_rf.fit(X_train, y_train)
    print(grid_search_rf.best_params_)
    print(grid_search_rf.best_score_)
```

## Adaboost

```python
abr = AdaBoostRegressor()
abr.fit(X_train, y_train)
```

```python
abr.score(X_test, y_test)
```

## Gradient Boosting

```python
param_grid_gb = {
    'loss': ['ls', 'lad', 'huber', 'quantile'],
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [50, 100, 150]
}
gbr = GradientBoostingRegressor()
```

```python
grid_search_gb = GridSearchCV(gbr, param_grid_gb, n_jobs=2)
for i in range(10):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xvals, yvals)
    grid_search_gb.fit(X_train, y_train)
    print(grid_search_gb.best_params_)
    print(grid_search_gb.best_score_)
```

```python
gbr = GradientBoostingRegressor(loss='huber', n_estimators=150)
gbr.fit(X_train, y_train)
```

```python
gbr.score(X_test, y_test)
```
